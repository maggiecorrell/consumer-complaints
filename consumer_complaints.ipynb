```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
%matplotlib inline
```

```python
dec_complaints = pd.read_csv('complaints_dec_2014.csv')
```

```python
product = dec_complaints['Product'].value_counts()
```

```python
product.plot(kind='bar', color = 'yellow')
plt.ylabel("number of complaints")
plt.xlabel("product")
plt.title("Number of complaints by product")
```

### Observations
The products with the highest number of complaints were mortgage and debt collection.  Followed by credit reporting.  Then a significant drop to complaints about bank account or service, credit card, consumer loan, student loan, payday loan, money transfers, prepaid card, and other financial services.

```python
company = dec_complaints['Company'].value_counts().head(10)
```

```python
company.plot(kind='bar', color = 'limegreen')
plt.ylabel("number of complaints")
plt.xlabel("company")
plt.title("Top 10 Companies")
```

### Observations
The chart plots the top 10 companies.  They start with the company with the largest number of complaints, Bank of America, and show a decrease as you go towards Capital one, that is in the 10th position. 

```python
company_response = dec_complaints['Company response'].value_counts()
```

```python
company_response.plot(kind='bar', color = 'orange')
plt.ylabel("number of complaints")
plt.xlabel("company response")
plt.title("Number of Complaints by Company Response")
```

### Observations
Shows the companies response to complaints filed.  By bar the highest response is closed with explanation.  Next highest company response, closed with non-monetary relief, sees a significant drop in response for complaint.  Lastly in progress, closed with monetary releif, closed, and untimely response are decreasing according to order.

```python
dec_complaints['day'] = pd.to_datetime(dec_complaints['Date received']).map(lambda d: d.weekday())
dec_complaints.groupby(['day', 'Date received']).size().groupby(level=0).mean().plot(kind = 'bar', color = 'gold')
plt.title("Mean number of complaints per day")
```

### Observation
The most complaints come in on week days.  Most noticably on Tuseday and Monday.  Saturday and Sunday have the least amount.
